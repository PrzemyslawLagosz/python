# 236. Feature Engineering


```python
label = df.median_house_value.copy()
label 
```

```python
features = df.drop(columns= "median_house_value")
features
```

```python
features.info()
```

```python
features.select_dtypes("float")
```

```python
import scipy.stats as stats
```

```python
features_norm = features.select_dtypes("float").apply(lambda x: stats.zscore(x))
features_norm
```

```python
pd.options.display.float_format = '{:.2f}'.format # Set displayiong float values to 2 decimal points
```

```python
features_norm.agg(["mean", "std"])
```

### Changing str to dummy variables

```python
features.ocean_proximity.value_counts()
```

## Uzywając modelu regresji liniowej `, drop_first= True` by wywalić niepotrzebną kolumne. W przypadku RANDOM FOREST nie ma znaczenia

```python
dummies = pd.get_dummies(features.ocean_proximity) # , drop_first= True
dummies
```

```python
features = pd.concat([features_norm, dummies, features.income_cat], axis=1)
features
```

# Test / Train

```python
TEST_SIZE = 0.2
```

```python
X_test = features.sample(frac= TEST_SIZE, random_state= 123)
```

```python
X_test
```

## median_income ma najwiekszy wpływ  na finalną cena. Upewnić się że wszystkie 5 kategori cenowych `income_cat` jest odpowiedno tyle samo jak w orginalnym secie. Przy random sampling moze wystapić sytuacja że np. wszystkie tanie mieszkania wylądują razem. `Stratified Sampling`.  


```python
X_test.income_cat.value_counts(normalize= True)
```

```python
features.income_cat.value_counts(normalize = True) 
```

## Kolejność indexów jest pomieszana `!`

```python
X_test.index
```

```python
X_train = features.loc[~features.index.isin(X_test.index)].copy()
```

```python
X_train
```

```python
X_train.income_cat.value_counts(normalize= True)
```

## Wymieszanie kolejnosci indexów `!`

```python
X_train = X_train.sample(frac= 1, random_state= 123)
X_train
```

```python
X_train.drop(columns = "income_cat", inplace= True)
X_test.drop(columns = "income_cat", inplace= True)
```

```python
y_train = label.loc[X_train.index]
y_test = label.loc[X_test.index]
```

```python
y_train # Zgadza się kolenosc z indexami w X_train
```

# 239. Training the Machine Learning Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
forest_reg = RandomForestRegressor(random_state= 42, n_estimators= 500, max_features= "sqrt", max_depth= 75, min_samples_split= 2)
```

```python
forest_reg.fit(X_train, y_train)
```

```python
forest_reg.score(X_train, y_train)
```

```python
pred = forest_reg.predict(X_train)
pred
```

```python
forest_mse = mean_squared_error(y_train, pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
forest_reg.score(X_train, pred)
```

# Evaluating the Model of the Test Set

```python
forest_reg.score(X_test, y_test)
```

```python
pred = forest_reg.predict(X_test)
pred
```

```python
forest_mse = mean_squared_error(y_test, pred)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
comp = pd.DataFrame(data= {"True_V": y_test, "Pred": pred})
comp
```

```python
ae = comp.True_V.sub(comp.Pred).abs()
ae
```

```python
mae = ae.mean()
mae
```
