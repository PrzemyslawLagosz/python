```python
plt.style.available
```

```python
xticks = [x for x in range(0,901, 50)]
```

```python
yticks = [x for x in range(0,81, 5)]
```

```python
plt.rcdefaults()
```

```python
titanic.age.plot(figsize= (12, 8), fontsize= 13, c = "red", linestyle = ".", 
                 xlim= (0, 900), ylim= (0, 80), xticks= xticks, yticks= yticks, rot = 45)
plt.title("Titanic - Ages", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize = (12,8), fontsize= 13, c = "r", linestyle = "-",
                 xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks, rot = 45) 
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
#plt.grid()
plt.show()
```

### Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts()
```

```python
titanic.age.value_counts(bins = [x for x in range (0, int(max(titanic.age)+1), 5)])#.sort_index()
```

```python
titanic.age.plot(kind= "line", figsize= (12, 8),  fontsize= 15)
plt.show()
```

```python
titanic.age.plot(kind = "hist", figsize = (12,8), fontsize = 15, bins = 80, density = False)
plt.show()
```

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize=15, ylabelsize= 15, cumulative = True)
plt.show()
```

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative= True)
plt.show()
```

### Scatterplots

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
type(titanic.survived)
```

```python
row = pd.Categorical(titanic['survived'], categories=[0, 1])
```

```python
titanic["new"] = row
```

```python
type(titanic.new)
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x = "age", y = "fare", c = "new", marker = "x", s = 20, colormap= "RdYlGn" ) 
plt.show()
```
